```python
import pandas as pd

spd_data = pd.read_csv("SPD.csv")
# 捐赠金额
amount = spd_data.groupby(by=['State'])['Amount'].sum()
amount = amount.sort_values(ascending = False)
state = list(amount.index)
state = [i.strip()for i in state]
data = [list(z) for z in zip(state, list(amount))]
max_amount = max(amount)
```

```python
# 捐赠数量
number = spd_data.groupby(by=['State'])['Amount'].count()
number = number.sort_values(ascending = False)
state_donate_number = list(number.index)
state_donate_number = [i.strip()for i in state_donate_number]
number_data = [list(z) for z in zip(state_donate_number, list(number))]
max_number = max(number)
print(number_data)
print(max_number)
```

```python
from pyecharts.charts import Pie   # Pie类用于绘制玫瑰图

# 接着导添加绘图的基本配置
pie = Pie().add("Donation Amount", # 添加提示框标签
                data, # 输入数据
                radius = ["20%", "80%"],  # 设置内半径和外半径
                center = ["60%", "60%"],  # 设置圆心位置
                rosetype = "radius")   # 玫瑰图模式，通过半径区分数值大小，角度大小表示占比

# 最后设置全局和系列配置项并绘制玫瑰图
pie.set_global_opts(title_opts = opts.TitleOpts(title="Rose chart of donations by state",  # 设置图标题
                                                pos_right = '60%'),  # 图标题的位置
                    legend_opts = opts.LegendOpts( # 设置图例
                                                orient='vertical', # 垂直放置图例
                                                pos_right="75%", # 设置图例位置
                                                pos_top="20%"))
pie.set_series_opts(label_opts = opts.LabelOpts(formatter="{b} : {d}%"))    
pie.load_javascript()
pie.render_notebook()
```
